```python
import pytesseract
from PIL import Image
import os
```

```python


# Load the image
img = Image.open('Moïse  Iloo Liandja.png')

```

```python
directory = f'pdf to images'
i = 0
for filename in os.listdir(directory):
    i = i + 1
    file = os.path.join(directory, filename)
    #checking if file exist
    if os.path.isfile(file):
        print(file, i)
        
        
        img = Image.open(file)
        text = pytesseract.image_to_string(img)
        
        file = file[0:-8]
        with open(f'{file}.txt', 'w') as file:
            file.write(text)
        file.close()
        break
        
```

```python
# Perform OCR
text = pytesseract.image_to_string(img)

```

```python
# Print the recognized text
print(text)
```

```python
# Define the directory path
dir_path = 'pdf to images/ocr'

# Get a list of all files in the directory
files = os.listdir(dir_path)

# Loop through each file in the directory
for file in files:
    # Check if the file ends with ".pdf.png.txt"
    if file.endswith('.pdf.png.txt'):
        # Generate the new file name by replacing ".pdf.png.txt" with ".txt"
        new_file_name = file.replace('.pdf.png.txt', '.txt')
        # Construct the full paths for the old and new file names
        old_path = os.path.join(dir_path, file)
        new_path = os.path.join(dir_path, new_file_name)
        # Rename the file
        os.rename(old_path, new_path)
        print(f'Renamed {file} to {new_file_name}')
```
